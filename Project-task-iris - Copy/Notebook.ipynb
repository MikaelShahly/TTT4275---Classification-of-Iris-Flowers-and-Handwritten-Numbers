```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import LogisticRegression as lr
```

```python
iris_df = pd.read_csv('iris.data', delimiter=',', header=None)
iris_df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
```

#### Lets preapre data-set to be used in classification model

```python
#encode the labels to be able to use in classification model
mapping = {
    'Iris-setosa' : 1,
    'Iris-versicolor' : 2,
    'Iris-virginica' : 3
    }
iris_df_categorical = iris_df.copy()
iris_df['class'] = iris_df['class'].map(mapping)
```

```python
#lets section out train / test set

#split train / test for task 1a)
iris_train_a_df = pd.concat([iris_df[0:30], iris_df[50:80], iris_df[100:130]])
iris_test_a_df = pd.concat([iris_df[30:50], iris_df[80:100], iris_df[130:150]])

iris_train_a_df.reset_index(drop=True, inplace=True)
iris_test_a_df.reset_index(drop=True, inplace=True)

X_train_a = iris_train_a_df.drop("class", axis=1)
X_test_a = iris_test_a_df.drop("class", axis=1)

y_train_a = iris_train_a_df["class"]
y_test_a = iris_test_a_df["class"]

#split datset for train/test for task d)
iris_train_d_df = pd.concat([iris_df[20:50], iris_df[70:100], iris_df[120:150]])
iris_test_d_df = pd.concat([iris_df[0:20], iris_df[50:70], iris_df[100:120]])

iris_train_d_df.reset_index(drop=True, inplace=True)
iris_test_d_df.reset_index(drop=True, inplace=True)

X_train_d = iris_train_d_df.drop("class", axis=1)
X_test_d = iris_test_d_df.drop("class", axis=1)

y_train_d = iris_train_d_df["class"]
y_test_d = iris_test_d_df["class"]
```

```python
def train_test_split(iris_df):
    """
    function for splitting data into a train / test set. 
    splits first 30 samples for each class into train
    and last 20 samples for each class into test

     Args:
            iris_df (pandas:dataFrame<m,n>): 
                pandas dataframe with our dataset

            
        Returns:
            dataset split into train and test datasets
    """
    iris_train_df = pd.concat([iris_df[0:30], iris_df[50:80], iris_df[100:130]])
    iris_test_df = pd.concat([iris_df[30:50], iris_df[80:100], iris_df[130:150]])

    iris_train_df.reset_index(drop=True, inplace=True)
    iris_test_df.reset_index(drop=True, inplace=True)

    X_train = iris_train_df.drop("class", axis=1)
    X_test = iris_test_df.drop("class", axis=1)
    y_train = iris_train_a_df["class"]
    y_test = iris_test_a_df["class"]

    return X_train, y_train, X_test, y_test
```

```python
def train_test_split_d(iris_df):
    """
    function for splitting data into a train / test set. 
    splits last 30 samples for each class into train
    and first 20 samples for each class into test

     Args:
            iris_df (pandas:dataFrame<m,n>): 
                pandas dataframe with our dataset

            
        Returns:
            dataset split into train and test datasets
    """
    iris_train_d_df = pd.concat([iris_df[30:60], iris_df[70:100], iris_df[120:150]])
    iris_test_d_df = pd.concat([iris_df[0:30], iris_df[50:70], iris_df[100:120]])

    iris_train_d_df.reset_index(drop=True, inplace=True)
    iris_test_d_df.reset_index(drop=True, inplace=True)

    X_train = iris_train_d_df.drop("class", axis=1)
    X_test = iris_test_d_df.drop("class", axis=1)

    y_train = iris_train_d_df["class"]
    y_test = iris_test_d_df["class"]

    return X_train, y_train, X_test, y_test
```

2c) 
Linear Seperability will be in the about seperability in the feature space. We might have linear seperability in certain dimensions of the feature space, but not in other dimensions. By linear seperability we mean classes can be seperated by a linear hyperplane.

By removing certain features where the classes are not linearly seperable (in the dimension), we can induce linear seperability as we now only fit the model in the dimensions where we have linear seperability. The model will then detect this linear seperability and be able to converge.

We are here doing feature selection in order to produce a linearly seperable feature space.
We could also have transformed the features, such as squaring them, in order to try to attain a linearly seperable feature space.
